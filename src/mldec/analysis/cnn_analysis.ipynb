```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
target = "cnn_toy_problem_only_good_examples_results_2.csv"
df = pd.read_csv(target)
```

```python
# histogram of 'val_acc' 
# df['val_acc'].hist(bins=20)
def compare_slice_by_column(df, column, slice_by):
    best = df[df[slice_by] == 1]
    cts, bins, _ = plt.hist(best[column], bins=20, alpha=1, color='r')
    plt.hist(df[column], bins=bins, alpha=0.5, color='b')
    plt.title(column)
    plt.show()

for column in ['lr', 'dropout', 'kernel_size', 'conv_channels', 'n_layers']:
    compare_slice_by_column(df, column, 'val_acc')
```
